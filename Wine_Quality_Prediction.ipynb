#Training a model for predicting the quality of wine using Support Vector Machine

**Data Source :-**

Data is downloaded from YBI foundation dataset github profile

###Importing Libraries :-

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn import svm
from sklearn.metrics import classification_report
```

###Reading data from the given data source

```python
data = pd.read_csv('https://github.com/YBI-Foundation/Dataset/raw/main/WhiteWineQuality.csv', sep = ';')
```

###Preparing the input and output data

```python
# input data is stored under feature_data
feature_data = data.drop('quality', axis = 1)
```

```python
# output data is stored under prediction_value
prediction_value = data['quality'].apply(lambda y_value: 1 if y_value >= 6 else 0)
```

Transforming input and output into numpy array



*   input under X
*   output under y



```python
X = np.asarray(feature_data)
y = np.asarray(prediction_value)
```

#**Train_Test_Split**




```python
# Splitting data into 70% training data and 30% testing data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, stratify=y, random_state=2932)
```

```python
# Storing the SVM model under classifier variable
classifier = svm.SVC(kernel='linear', gamma='auto', C=2)
```

```python
# Training the model using fit function
classifier.fit(X_train, y_train)
```

```python
# Storing the predicting results of testing data under y_predict
y_predict = classifier.predict(X_test)
```

#**Results**

```python
# demonstrating the metrics of the prediction
print(classification_report(y_test, y_predict))
```
